```python
import altair as alt
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_bcubcg_fall2022/main/data/bfro_reports_fall2022.csv')
df.head()
```

```python
df.columns
```

```python
df_first_plot = df[['state','temperature_mid', 'humidity', 'precip_type']]
df_first_plot.head()
```

```python
df_first_plot['precip_type'].unique()
```

```python
df_first_plot['temperature_mid'].unique()
```

```python
df_first_plot['humidity'].unique()
```

```python
df_first_plot.isna().sum()
```

```python
df_first_plot = df_first_plot.dropna()
df_first_plot.head()
```

```python
first_plot = alt.Chart(df_first_plot).mark_circle().encode(
    alt.X("humidity:Q"),
    alt.Y("temperature_mid:Q"),
    alt.Color("precip_type:N"),
    tooltip=['state', 'humidity', 'temperature_mid', 'precip_type']
).properties(
    width=600,
    height=400
).interactive()
first_plot
```

```python
first_plot.properties(width='container').save('HW10_first_plot.json')
```

```python
df_second_plot = df[['cloud_cover', 'visibility', 'wind_speed', 'summary', 'season']]
df_second_plot.head()
```

```python
df_second_plot.isna().sum()
```

```python
df_second_plot = df_second_plot.dropna()
df_second_plot.head()
```

```python
df_second_plot['season'].unique()
```

```python
df_second_plot = df_second_plot[df_second_plot['season'].str.contains('Unknown') == False]
```

```python
df_second_plot['season'].unique()
```

```python
second_plot = alt.Chart(df_second_plot).mark_circle().encode(
    alt.X("cloud_cover:Q"),
    alt.Y("visibility:Q"),
    alt.Color("season"),
    size='wind_speed',
    tooltip=['season', 'cloud_cover', 'wind_speed', 'visibility', 'summary']
).properties(
    width=800,
    height=400
).interactive()
second_plot
```

```python
second_plot.properties(width='container').save('HW10_second_plot.json')
```
